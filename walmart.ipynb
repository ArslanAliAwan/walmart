```python
def classification_model(model,sample_dataset,prediction_input,output):
    model.fit(sample_dataset[prediction_input],sample_dataset[output]) #Here we fit the model using training set
    predictions = model.predict(sample_dataset[prediction_input])
    accuracy = metrics.accuracy_score(predictions,sample_dataset[output])
    print("Accuracy : %s" % "{0:.3%}".format(accuracy))
 
    
    kf = KFold(sample_dataset.shape[0], n_folds=5)

    error = []
    for train, test in kf:

        train_X = (sample_dataset[prediction_input].iloc[train,:])
        train_y = sample_dataset[output].iloc[train]
        model.fit(train_X, train_y)
    
        
        test_X=sample_dataset[prediction_input].iloc[test,:]
        test_y=sample_dataset[output].iloc[test]
        error.append(model.score(test_X,test_y))
        
        print("Cross-Validation Score : %s" % "{0:.3%}".format(np.mean(error)))
```
